```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [i for i in range(1, 8)]
y = [3.5, 3.8, 4.2, 4.5, 5, 5.5, 7]
x, y
```

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080', 'family': 'Calibri'}
label_dict = {'fontsize': '14', 'color': '#808080', 'family': 'Calibri'}
legend_dict = {'family': 'Calibri', 'size': 10}
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)

ax[0].plot(x, y)
ax[0].set_title('first subplot')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')

ax[1].scatter(x, y)
ax[1].set_title('second subplot')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')

plt.subplots_adjust(hspace=0.6)
plt.show()
```
